#**Translation**

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

src = cv2.imread(r'D:\AIDS_PR_EXAM\Computer Vision\EddieG.jpg')
img = cv2.cvtColor(src, cv2.COLOR_BGR2RGB)
```

```python
M = np.array([[1, 0, 50], [0, 1, 0]], dtype=np.float32)
translated_x = cv2.warpAffine(img, M, (img.shape[1], img.shape[0]))
plt.imshow(translated_x)
```

```python
M = np.array([[1, 0, 0], [0, 1, 50]], dtype=np.float32)
translated_y = cv2.warpAffine(img, M, (img.shape[1], img.shape[0]))
plt.imshow(translated_y)
```

```python
M = np.array([[1, 0, 50], [0, 1, 50]], dtype=np.float32)
translated_both = cv2.warpAffine(img, M, (img.shape[1], img.shape[0]))
plt.imshow(translated_both)
#both x & y
```

#**Reflection**

```python
#Reflection horizontal
M = np.array([[1, 0, 0], [0, -1, img.shape[0]]], dtype=np.float32)
reflected_horizontal = cv2.warpAffine(img, M, (img.shape[1], img.shape[0]))
plt.imshow(reflected_horizontal)
```

```python
#Reflection vertical
M = np.array([[-1, 0, img.shape[1]], [0, 1, 0]], dtype=np.float32)
reflected_vertical = cv2.warpAffine(img, M, (img.shape[1], img.shape[0]))
plt.imshow(reflected_vertical)
```

```python
M = np.array([[-1, 0, img.shape[1]], [0, -1, img.shape[0]]], dtype=np.float32)
reflected_both = cv2.warpAffine(img, M, (img.shape[1], img.shape[0]))
plt.imshow(reflected_both)
```

# **Rotation**

```python
center = (img.shape[1] // 2, img.shape[0] // 2)
M = cv2.getRotationMatrix2D(center, 45, 1)
rotated_45 = cv2.warpAffine(img, M, (img.shape[1], img.shape[0]))
plt.imshow(rotated_45)
```

```python
M = cv2.getRotationMatrix2D(center, 90, 1)
rotated_90 = cv2.warpAffine(img, M, (img.shape[1], img.shape[0]))
plt.imshow(rotated_90)
```

```python
M = cv2.getRotationMatrix2D(center, -60, 1)
rotated_minus_60 = cv2.warpAffine(img, M, (img.shape[1], img.shape[0]))
plt.imshow(rotated_minus_60)
```

# **Scaling**

```python
M = np.array([[1.5, 0, 0], [0, 1.5, 0]], dtype=np.float32)
scaled_img = cv2.warpAffine(img, M, (int(img.shape[1] * 1.5), int(img.shape[0] * 1.5)))
plt.imshow(scaled_img)
```

```python
M = np.array([[0.5, 0, 0], [0, 0.5, 0]], dtype=np.float32)
scaled_img = cv2.warpAffine(img, M, (int(img.shape[1] * 0.5), int(img.shape[0] * 0.5)))
plt.imshow(scaled_img)
```

# **Shearing**

```python
M = np.array([[1, 0.5, 0], [0, 1, 0]], dtype=np.float32)
sheared_x = cv2.warpAffine(img, M, (int(img.shape[1] + 0.5 * img.shape[0]), img.shape[0]))
plt.imshow(sheared_x)
```

```python
M = np.array([[1, 0, 0], [0.5, 1, 0]], dtype=np.float32)
sheared_y = cv2.warpAffine(img, M, (img.shape[1], int(img.shape[0] + 0.5 * img.shape[1])))
plt.imshow(sheared_y)
```

```python
M = np.array([[1, 0.3, 0], [0.3, 1, 0]], dtype=np.float32)
sheared_both = cv2.warpAffine(img, M, (int(img.shape[1] + 0.3 * img.shape[0]), int(img.shape[0] + 0.3 * img.shape[1])))
plt.imshow(sheared_both)
```

# **Combined Effect**

```python
center = (img.shape[1] // 2, img.shape[0] // 2)
M_rot = cv2.getRotationMatrix2D(center, 45, 1)
rotated_img = cv2.warpAffine(img, M_rot, (img.shape[1], img.shape[0]))

M_trans = np.array([[1, 0, 50], [0, 1, 50]], dtype=np.float32)
rotated_translated = cv2.warpAffine(rotated_img, M_trans, (rotated_img.shape[1], rotated_img.shape[0]))
plt.imshow(rotated_translated)
```

```python
M_trans = np.array([[1, 0, 50], [0, 1, 50]], dtype=np.float32)
translated_img = cv2.warpAffine(img, M_trans, (img.shape[1], img.shape[0]))

M_rot = cv2.getRotationMatrix2D((translated_img.shape[1] // 2, translated_img.shape[0] // 2), 45, 1)
translated_rotated = cv2.warpAffine(translated_img, M_rot, (translated_img.shape[1], translated_img.shape[0]))
plt.imshow(translated_rotated)
```
