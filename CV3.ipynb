```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread(r'D:\AIDS_PR_EXAM\Computer Vision\ReyMysterio.jpg')
image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image, cmap='gray')
plt.show()
```

## Same KernelSize, Different Sigma

```python
gblur1 = cv2.GaussianBlur(gray_image, (11,11), 0.5)
plt.imshow(gblur1, cmap='gray')
plt.show()
```

```python
gblur2 = cv2.GaussianBlur(gray_image, (11,11), 1)
plt.imshow(gblur2, cmap='gray')
plt.show()
```

```python
gdiff = gblur1 - gblur2
plt.imshow(gdiff, cmap='gray')
plt.show()
```

```python
gdiff2 = gblur2 - gblur1
plt.imshow(gdiff2, cmap='gray')
plt.show()
```

## Same Sigma, Different KernelSize

```python
gaus1 = cv2.GaussianBlur(gray_image, (5,5), 1)
plt.imshow(gaus1, cmap='gray')
plt.show()
```

```python
gaus2 = cv2.GaussianBlur(gray_image, (7,7), 1)
plt.imshow(gaus2, cmap='gray')
plt.show()
```

```python
gausdiff1 = gaus1 - gaus2
plt.imshow(gausdiff1, cmap='gray')
plt.show()
```

```python
gausdiff2 = gaus2 - gaus1
plt.imshow(gausdiff2, cmap='gray')
plt.show()
```

```python
gausdiff2.nonzero()
```

## Threshold

```python
np.unique(gausdiff2)
```

```python
threshold2 = gaus2 - gaus1
for y in range(threshold2.shape[0]):
    for x in range(threshold2.shape[1]):
        if threshold2[y,x]<1:
            threshold2[y,x] = 0
        else:
            threshold2[y,x] = 255

plt.imshow(threshold2, cmap='gray')
plt.show()
```
